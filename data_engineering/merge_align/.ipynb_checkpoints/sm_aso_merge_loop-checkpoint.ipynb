# Description
This code runs functions to reproject and align aso geotiffs with SnowModel output

```python
"""
    LOAD PYTHON PACKAGES
"""
import xarray as xr
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
import os
import pyproj
import rioxarray as rxr
import datetime
```

```python
"""
    LOAD FUNCTIONS
"""
from process.geography import sm_projected_coordinates as sm_proj
from process.sm_aso import sm_aso_match as grid_match
```

```python
"""
    INPUTS
"""
sm_fpath = '/glade/scratch/rossamower/snow/snowmodel/aso/california/tuolumne/snowmodel/cheyenne/outputs/'
water_yr_lst = ['2013','2014','2015','2016','2017','2018','2019','2020']
proj_string = '+proj=utm +zone=11 +ellps=WGS84 +datum=WGS84 +units=m +no_defs'
crs = 32611
aso_fpath = '/glade/u/home/rossamower/work/git_misc/Tuol_ASO/data/aso/'
```

```python
"""
    GET FIXED SNOWMODEL PARAMETERS [lat,lon,sm_x,sm_y,veg,hgt]
"""
## create projection object ##
sm_proj = sm_proj(water_yr_lst[0],sm_fpath,proj_string)

## call reproject method ##
sm_proj.reproject()

## assign snowmodel variables ##
lat = sm_proj.lat
lon = sm_proj.lon
vgt = sm_proj.vgt
hgt = sm_proj.hgt
sm_x = sm_proj.sm_x
sm_y = sm_proj.sm_y
```

```python
"""
    RUN LOOP TO MATCH ASO AND SM
"""
obj = grid_match(sm_fpath,aso_fpath,water_yr_lst,
                 sm_x,sm_y,vgt,hgt,crs)
```

```python
obj.loop()
```

```python
"""
    FINALIZE DATASET
"""
## save dataset ##
final_ds = obj.ds_concat

## add hgt to dataset ##
final_ds['hgt'] = xr.DataArray(data = obj.hgt,
                               dims=["y", "x"],
                               coords={
                            "y": (("y"),final_ds.y),
                            "x": (("x"),final_ds.x)})
## add veg to dataset ##
final_ds['veg'] = xr.DataArray(data = obj.veg,
                               dims=["y", "x"],
                               coords={
                            "y": (("y"),final_ds.y),
                            "x": (("x"),final_ds.x)})

```

```python
"""
    SAVE TO NETCDF
"""
ds_output_dir = '/glade/scratch/rossamower/snow/snowmodel/aso/california/tuolumne/snowmodel/cheyenne/outputs/aso_sm/'
final_ds.to_netcdf(ds_output_dir + 'aso_sm_merge.nc')
```
