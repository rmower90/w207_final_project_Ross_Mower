```python
"""
    RUNNING MASK - 0's and -9999.0
"""
import random
from sklearn.linear_model import LinearRegression
import math
from IPython.display import clear_output

pd_time = pd.DatetimeIndex(time)

mask_lst = []
count = 0
for i in range(0,ds.x.shape[0]):
    print(i,end = ' ')
    
    for j in range(0,ds.y.shape[0]):  
        y_test = ds.aso_swe[:,j,i].values
        x_test = ds.sm_swe[:,j,i].values
        date_test_date = np.array(pd.to_datetime(ds.time[:].values))
        date_test_month = np.array(pd.to_datetime(ds.time[:].values).month)
        date_test_year = np.array(pd.to_datetime(ds.time[:].values).year)
        
        for k in range(0,len(y_test)):
            count += 1
            if np.isnan(y_test[k]) == True:
                ## test for nans ##
                mask_lst.append(1)
                
            if (y_test[k] < 0.0):
                ## test for -9999.0 ##
                mask_lst.append(1)
                
            if (y_test[k] > 0.0):
                mask_lst.append(0)
                
            if (y_test[k] == 0.0):
                if (k > 0) and (k < len(y_test)-1):
                    if (date_test_year[k-1] == date_test_year[k]) and (date_test_year[k+1] == date_test_year[k]):
                        if (y_test[k-1] > 0.05) and (y_test[k+1] > 0.05):
                            mask_lst.append(1)
                        else:
                            mask_lst.append(0)
                    else:
                        mask_lst.append(0)
                else:
                    mask_lst.append(0)
                    
        if (len(mask_lst) != count):
            print('DATA MISMATCH!')
            input('PLEASE ENTER VALUE TO ACKNOWLEDGE PROBLEM')
                
```

```python
"""
    TESTING MASK
"""
mask_1d = np.array(mask_lst)
mask_3d = mask_1d.reshape(ds.time.shape[0],ds.y.shape[0],ds.x.shape[0],order = 'F')
aso_after = np.where(mask_3d == 1, np.nan, ds.aso_swe.values)

## test values changed ##
j = 589
i = 805
# j = 600
# i = 600
# j = 791
# i = 992
# j = 48
# i = 1145
# j = 825
# i = 1320

print('OLD')
print(ds.aso_swe[:,j,i].values)
print()
print('NEW')
print(mask_3d[:,j,i])
print(aso_after[:,j,i])
```

```python
"""
    MASK FOR SUFFICIENT SAMPLES IN GRID ANALYSIS
"""
import random
from IPython.display import clear_output

yr_list = [[2013,2014],[2019,2020]]

j = random.randint(0,ds.y.shape[0])
i = random.randint(0,ds.x.shape[0])

count_valid = 0
count_lst = []

## Loop through grid cells and identify whether there are at least 5 (non-nan) test samples and 30 (non-nan) training samples for each grid cell ##
### create a  mask to be used for the grid analysis ##

for i in range(0,ds.x.shape[0]):
    if i%10 == 0:
        print(i,end = ' ')
    for j in range(0,ds.y.shape[0]):

        y_val = aso_after[:,j,i]
        date_month = np.array(pd.to_datetime(ds.time[:].values).month)
        date_year = np.array(pd.to_datetime(ds.time[:].values).year)
        d = {}
        isnan = False
        
        for yr in yr_list:
            # training #
            train_bool = [x not in yr for x in pd_time.year]
            y_train = y_val[train_bool]
            y_train_nonnan = len(y_train) - np.isnan(y_train).sum()
            
            # testing #
            test_bool = [x in yr for x in pd_time.year]
            y_test = y_val[test_bool]
            y_test_nonnan = len(y_test) - np.isnan(y_test).sum()
            
            if (y_test_nonnan <= 5) or (y_train_nonnan < 30):
                isnan = True
        
        if isnan == False:
            count_valid += 1
            count_lst.append(0)
        else:
            count_lst.append(1)
            
print(f'Valid count :{count_valid}')
print(f'Invalid count :{(ds.x.shape[0] * ds.y.shape[0]) - count_valid}')


mask_1d = np.array(count_lst)
not_grid = mask_1d.reshape(ds.y.shape[0],ds.x.shape[0],order = 'F')
```

```python
hgt = ds_trim.hgt.values 
asp_x_slp = ds_trim.asp_x_slp.values 

def bin_array(arry,ds):
    count_, bin_ = np.histogram(arry.flatten(),bins = 10)
    hot = np.digitize(arry.flatten(), bin_, right=False)
    hot = hot.reshape(ds.y.shape[0],ds.x.shape[1])

    ## name mapping ##
    elev_dict = {}
    for val in range(0,len(bin_)):
        if val == 0:
            elev_dict[val] = f'0.0-{np.floor(bin_[val])}'
        else:
            elev_dict[val] = f'{np.ceil(bin_[val-1])}-{np.floor(bin_[val])}'
            
    print('')
            
    for k,v in elev_dict.items():
        print(f'{k} : {v}')
        
    return hot, elev_dict


elev_hot, elev_hot_vals = bin_array(hgt,ds_trim)
asp_x_slp_hot, asp_x_slp_vals = bin_array(asp_x_slp,ds_trim)
```
